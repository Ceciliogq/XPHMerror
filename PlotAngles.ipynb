```python
def XPHMAngles(filename):
    file = open(filename,'r')
    lines=file.readlines()[2:]
    result=[]
    for x in lines:
        result.append([float(item) for item in x.split()]) 
    file.close()
    
    result = np.array(result)
    freqs = result[:,0]
    alpha = result[:,1]
    epsilon = result[:,2]
    beta = result[:,3]
    
    figure, axes = plt.subplots(nrows=1, ncols=3, figsize=(20,4))    
    axes[0].plot(freqs, alpha, label='alpha')
    axes[1].plot(freqs, epsilon, label='epsilon')
    axes[2].plot(freqs, beta, label='beta')
    axes[0].legend()
    axes[1].legend()
    axes[2].legend()
    axes[0].set_ylim(-2, 1.1*max(alpha))
    axes[1].set_ylim(-2, 1.1*max(epsilon))
    axes[0].set_xlim(10,300)
    axes[1].set_xlim(10,300)
    axes[2].set_xlim(10,300)
    figure.suptitle(filename)
    plt.show()
```

```python
mode_array = ['22','21','32','33','44']
mode_array = ['21','32']
mode_array = ['22','21','32']
```

```python
# Coarse angles (no multibanding)
for mode in mode_array:
    XPHMAngles('coarse_angles_hphc_MB_'+mode+'.dat')
```

```python
# Fine angles with Multibanding
for mode in mode_array:
    XPHMAngles('angles_hphc_MB_'+mode+'.dat')
```

```python
# Fine Angles without Multibanding 
for mode in mode_array:
    XPHMAngles('angles_hphc_'+mode+'.dat')
```
